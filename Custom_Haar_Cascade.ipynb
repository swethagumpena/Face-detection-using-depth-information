### **CUSTOM HAAR CASCADE CLASSIFIER**
**First Step**: is to extract the Haar features                                       
**Second Step**: is integral image - These are basically mathematical computations to fasten the process.                                                  
**Third Step**: is Adaboost Classifier - This extracts the best features and rejects all the unnecessary features. It selects only the relevant features.          
**Fourth Step**: The job of each stage is to determine whether a given subwindow is definitely not a face or maybe a face. A given subwindow is immediately discarded if it fails to find a face in any of the stages.


```python
import numpy as np
import cv2
import time
```

```python
from google.colab.patches import cv2_imshow
```

```python
face_cascade=cv2.CascadeClassifier('/content/cascade (1).xml')
```

```python
img=cv2.imread('/content/KinectScreenshot-Depth-08-06-31.bmp')
#img=cv2.imread('/content/KinectScreenshot-Depth-06-38-45.bmp')
#img=cv2.imread('depth_10.png')

#img=cv2.imread('v4.jpeg')
#img=cv2.imread('negtest.jpg')
#img=cv2.imread('depth_84.png')
#img=cv2.imread('depth_white_10.png')
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
faces=face_cascade.detectMultiScale(gray,1.01,7)
cv2_imshow(img)
cv2.waitKey(0)
```

```python
for (x,y,w,h) in faces:
    img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
```

```python
start_time = time.time()
for (x,y,w,h) in faces:
    img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)

cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()



#main()
print("--- %s seconds ---" % (time.time() - start_time))
```
